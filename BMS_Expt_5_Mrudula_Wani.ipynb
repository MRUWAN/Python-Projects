```python
#Mrudula Wani
#22105A0031
#Write a program to simulate GHK equation
import numpy as np
import matplotlib.pyplot as plt

def ghk_equation(concentration_in, concentration_out, permeability):
    """
    Calculate the membrane potential using the Goldman-Hodgkin-Katz (GHK) equation.

    Parameters:
        concentration_in (float): Intracellular concentration of the ion (in mM).
        concentration_out (float): Extracellular concentration of the ion (in mM).
        permeability (float): Permeability of the ion.

    Returns:
        membrane_potential (float): The calculated membrane potential (in mV).
    """
    R = 8.314       # Gas constant (J/(K*mol))
    T = 293         # Temperature (K)
    F = 96485       # Faraday constant (C/mol)
    z = 1           # Valency of the ion (assuming monovalent)

    concentration_ratio = concentration_out / concentration_in
    membrane_potential = (R * T / (z * F)) * np.log(permeability * concentration_ratio)

    return membrane_potential

def main():
    ion_name = input("Enter the name of the ion: ")
    concentration_in = float(input(f"Enter the intracellular concentration of {ion_name} (in mM): "))
    concentration_out = float(input(f"Enter the extracellular concentration of {ion_name} (in mM): "))
    permeability = float(input(f"Enter the permeability of {ion_name}: "))

    membrane_potential = ghk_equation(concentration_in, concentration_out, permeability)
    print(f"The membrane potential for {ion_name} is {membrane_potential:.2f} mV")

    # Plotting
    concentrations_out_range = np.linspace(0, 200, 100)
    membrane_potentials = [ghk_equation(concentration_in, c_out, permeability) for c_out in concentrations_out_range]

    plt.plot(concentrations_out_range, membrane_potentials)
    plt.xlabel('Extracellular Concentration (mM)')
    plt.ylabel('Membrane Potential (mV)')
    plt.title(f'Membrane Potential vs Extracellular Concentration of {ion_name}')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
