```python

import matplotlib.pyplot as plt

# Sample data
year= [2001,2002,2003, 2004,2005]
cancer_patient= [10, 12, 8, 15, 11]

# Create a line chart
plt.plot(year,cancer_patient)

# Add labels and a title
plt.xlabel('year')
plt.ylabel('cancer_patient')
plt.title('Line Chart')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
patients = ['patient1', 'patient2', 'patient3', 'patient4']
Blood_pressure = [15, 30, 45, 10]

# Create a bar chart
plt.bar(patients,Blood_pressure  )

# Add labels and title
plt.xlabel('Patient')
plt.ylabel('Blood_Pressure ')
plt.title('Bar Chart Example')

# Display the chart
plt.show()
```
