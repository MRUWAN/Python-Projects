```python
import pandas as pd

# Create a sample dataset (you can replace this with your own dataset)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 22, 35, 28],
    'Blood_Glucose_Level': [100, 120, 141, 145, 150]
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())

# Basic statistics of the numeric columns
print(df.describe())

# Grouping and aggregating data
grouped = df.groupby('Age').mean()
print(grouped)

# Data visualization (you may need to install matplotlib or seaborn)
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x='Age', y='Blood_Glucose_Level')
plt.title('Age vs. Blood_Glucose_Level')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample medical history dataset
data = {
    'Patient_ID': [1, 2, 3, 4, 5],
    'Age': [35, 45, 28, 50, 42],
    'Blood_Pressure': [100, 140, 108, 100, 128],
    'Cholesterol': [180, 210, 190, 220, 200]
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Visualize age distribution
plt.figure(figsize=(8, 7))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age vs. BloodPressure')
plt.xlabel('Age')
plt.ylabel('BloodPressure')
plt.show()

# Visualize Blood Pressure vs. Cholesterol
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Cholesterol'], c='green', marker='o', alpha=0.8)
plt.title('Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```
